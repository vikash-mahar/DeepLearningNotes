```python
from sklearn.neighbors import KNeighborsClassifier

X=[
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [350,9]
]
y=[0,0,1,1,1] #0=apple , 1=orange

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
weight = float(input("enter weights"))
size = float(input("enter size"))
prediction = model.predict([[weight,size]])[0]

if prediction ==0:
    print("apple")
else:
    print("orange")
```
