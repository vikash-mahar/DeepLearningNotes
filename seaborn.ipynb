```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mart=pd.read_excel("/Users/vikasmahar/Desktop/AI:ML/SEABORN/sales.xlsx")
```

```python

#COUNT PLOT


mart.head(5)
plt.figure(figsize=(15,4))
# sns.countplot(x="Product line", hue="Gender", data=mart, palette='gist_grey')
sns.countplot(x='Product line',data=mart, facecolor= (1,0,1,0), linewidth=5, edgecolor=sns.color_palette('dark',3))


```

```python

#BAR PLOT

plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,
            order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel'],
            hue_order=['Male','Female'],capsize=0.1,color='red',palette='cividis',saturation=0.1,estimator=np.median)
#sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel'],hue_order=['Male','Female'],ci=None)
x=mart['Product line'].sort_values()
x.unique()

```

```python

#BOX PLOT

sns.boxplot(x='Payment', y='Total', hue='Gender', data=mart, showmeans= True, 
            meanprops={"marker":"o", "markerfacecolor":"white", "markersize":"10", "markeredgecolor":"black"})

#sns.boxplot(data=mart) box plot for each of the numeric variable in the dataframe
```

```python

#VIOLIN PLOT


sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, split=True, inner='quartile',cut=0)
#inner='stick'
```
