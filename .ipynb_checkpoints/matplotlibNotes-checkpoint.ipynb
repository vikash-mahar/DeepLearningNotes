```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[12,15,22,5,34]

plt.plot(x,y)
plt.show()
```

```python
x=['mon','tues','wed','thurs','fri']
y=[10,13,24,19,10]
plt.plot(x,y)
plt.title('Bakery sales thus week')
plt.xlabel('days')
plt.ylabel('sales')
plt.show()
```

```python
months=[1,2,3,4]
sales=[1000,1300,2000,1200]
plt.plot(months, sales, color='blue', linestyle='--', linewidth=2, marker='o', label ='2025 sales data')
plt.xlabel('months')
plt.ylabel('sales per month')
plt.title('monthly sales data report')
plt.legend(loc='upper left', fontsize=12)
plt.grid(color='gray', linestyle=':', linewidth=1)
plt.xlim(1,5)
plt.ylim(0,2400)
plt.xticks([1,2,3,4,5],['m1','m2','m3','m4','m5'])
plt.show()
          
          
```

```python
hours=[1,2,3,4,5]
exams=[50,55,60,70,75]
plt.scatter(hours, exams, color='green', marker='o', label='student data')
plt.xlabel("hours studied")
plt.ylabel("exams Score")
plt.title("relationship between study time and exam score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3], [50,55,60], color='green', label='class A')
plt.scatter([1,2,3], [54,55,65], color='orange', label='class B')
plt.xlabel("hours studied")
plt.ylabel("exams Score")
plt.title("relationship between study time and exam score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,15,24]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')

plt.show()
```

```python
fig, ax= plt.subplots(1,2 , figsie=(10,5))
x=[1,2,3,4]
y=[10,20,15,24]

ax[0].plot(x,y)
ax[0].set_title('line plot')

ax[1].bar(x,y, color='green')
ax[1].set_title('bar plot')
fig, ax= plt.subplots(2,1, sharex= True)
fig.suptitle('comparision of line and bar chart')

plt.tight_layout()
plt.show()
```
